# Dependent Random Variables

## Overview

In this notebook, we will explore the concept of **dependent random variables** using a contingency table. Dependent random variables are variables whose outcomes are not independent of each other. This means that the value of one variable can influence the probability distribution of the other.

### We will first consider the following concepts:

1. **Contingency Tables**:
   - A contingency table is a matrix that shows the joint distribution of two categorical random variables.
   - From the table, we can calculate the **marginal probabilities** of each variable.

2. **Covariance**:
   - Covariance is a measure of the relationship between two random variables. It tells us whether the variables tend to increase or decrease together.
   - A positive covariance indicates that the variables tend to increase together, while a negative covariance indicates that one variable tends to increase as the other decreases.

## What the next cells will cover:

1. **Step 1**: Start with a contingency table containing four numbers that represent the joint probabilities of two random variables, $X$ and $Y$.
2. **Step 2**: Calculate the **marginal probabilities** of $X$ and $Y$ from the contingency table.
3. **Step 3**: Compute the **covariance** between $X$ and $Y$ manually using the formula:
   $$
   \text{Cov}(X, Y) = \mathbb{E}[XY] - \mathbb{E}[X]\mathbb{E}[Y]
   $$
   where:
   - $\mathbb{E}[XY]$ is the expected value of the product of $X$ and $Y$.
   - $\mathbb{E}[X]$ and $\mathbb{E}[Y]$ are the expected values of $X$ and $Y$, respectively.

4. **Step 4**: Interpret the results to determine whether $X$ and $Y$ are dependent.

By the end of this notebook, you will have a better understanding of how to analyze the relationship between two random variables using contingency tables and covariance.

```python
# Step 1: Define the contingency table
import numpy as np
import pandas as pd

# Contingency table (joint probabilities of X and Y)
# Rows represent X values, columns represent Y values
contingency_table = np.array([
    [0.1, 0.2],  # P(X=0, Y=0) and P(X=0, Y=1)
    [0.3, 0.4]   # P(X=1, Y=0) and P(X=1, Y=1)
])

# Display the contingency table
df = pd.DataFrame(contingency_table, columns=["Y=0", "Y=1"], index=["X=0", "X=1"])
print("Contingency Table:")
display(df)
```

```python
# Step 2: Calculate marginal probabilities
# Marginal probabilities for X
marginal_X = np.sum(contingency_table, axis=1)  # Sum over rows
print("\nMarginal Probabilities for X:")
print(f"P(X=0) = {marginal_X[0]:.2f}, P(X=1) = {marginal_X[1]:.2f}")

# Marginal probabilities for Y
marginal_Y = np.sum(contingency_table, axis=0)  # Sum over columns
print("\nMarginal Probabilities for Y:")
print(f"P(Y=0) = {marginal_Y[0]:.2f}, P(Y=1) = {marginal_Y[1]:.2f}")

```

```python
# Step 3: Calculate covariance
# Expected value of X
E_X = np.sum([0, 1] * marginal_X)
print(f"\nE[X] = {E_X:.2f}")

# Expected value of Y
E_Y = np.sum([0, 1] * marginal_Y)
print(f"E[Y] = {E_Y:.2f}")

# Expected value of XY (see equation from the lecture)
E_XY = 0
for i, x in enumerate([0, 1]):  # Loop over X values
    for j, y in enumerate([0, 1]):  # Loop over Y values
        E_XY += x * y * contingency_table[i, j]
print(f"E[XY] = {E_XY:.2f}")

# Covariance
cov_XY = E_XY - (E_X * E_Y)
print(f"Cov(X, Y) = {cov_XY:.2f}")

# Step 4: Interpretation
if cov_XY > 0:
    print("\nThe covariance is positive, indicating that X and Y tend to increase together.")
elif cov_XY < 0:
    print("\nThe covariance is negative, indicating that X and Y tend to move in opposite directions.")
else:
    print("\nThe covariance is zero, indicating no linear relationship between X and Y.")
```

```python
################################################
#
#   S O L U T I O N !!
#

# Step 1: Construct the contingency table
# Rows represent gene expression (X), columns represent protein production (Y)
contingency_table = np.array([
    [40, 10],  # X=1 (Gene expressed): Y=1 (Protein produced), Y=0 (Protein not produced)
    [5, 45]    # X=0 (Gene not expressed): Y=1 (Protein produced), Y=0 (Protein not produced)
])

# Display the contingency table
df = pd.DataFrame(contingency_table, columns=["Y=1 (Protein Produced)", "Y=0 (Protein Not Produced)"],
                  index=["X=1 (Gene Expressed)", "X=0 (Gene Not Expressed)"])
print("Contingency Table:")
display(df)

# Step 2: Calculate marginal probabilities
total_cells = np.sum(contingency_table)  # Total number of cells (100)

# Marginal probabilities for X
marginal_X = np.sum(contingency_table, axis=1) / total_cells  # Sum over rows
print("\nMarginal Probabilities for X:")
print(f"P(X=1) = {marginal_X[0]:.2f}, P(X=0) = {marginal_X[1]:.2f}")

# Marginal probabilities for Y
marginal_Y = np.sum(contingency_table, axis=0) / total_cells  # Sum over columns
print("\nMarginal Probabilities for Y:")
print(f"P(Y=1) = {marginal_Y[0]:.2f}, P(Y=0) = {marginal_Y[1]:.2f}")

# Step 3: Calculate expected values
# Expected value of X
E_X = np.sum([0, 1] * marginal_X)
print(f"\nE[X] = {E_X:.2f}")

# Expected value of Y
E_Y = np.sum([0, 1] * marginal_Y)
print(f"E[Y] = {E_Y:.2f}")

# Expected value of XY
E_XY = 0
for i, x in enumerate([0, 1]):  # Loop over X values
    for j, y in enumerate([0, 1]):  # Loop over Y values
        E_XY += x * y * (contingency_table[i, j] / total_cells)
print(f"E[XY] = {E_XY:.2f}")

# Step 4: Calculate covariance
cov_XY = E_XY - (E_X * E_Y)
print(f"Cov(X, Y) = {cov_XY:.2f}")

# Step 5: Interpretation
if cov_XY > 0:
    print("\nThe covariance is positive, indicating that gene expression and protein production are positively correlated.")
elif cov_XY < 0:
    print("\nThe covariance is negative, indicating that gene expression and protein production are negatively correlated.")
else:
    print("\nThe covariance is zero, indicating no linear relationship between gene expression and protein production.")
```

## Calculating the Correlation Coefficient

The correlation coefficient is a measure of the linear relationship between two variables, $X$ and $Y$. It is calculated using the formula:

$$
\text{Correlation}(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$

Where:
- $\text{Cov}(X, Y)$ is the **covariance** between $X$ and $Y$.
- $\sigma_X$ is the **standard deviation** of $X$.
- $\sigma_Y$ is the **standard deviation** of $Y$.


   - The covariance is given by:
     $$
     \text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{X})(y_i - \bar{Y})
     $$
   - This measures how $X$ and $Y$ vary together.

   - The standard deviation of $X$ is:
  $$ \sigma_X = \sqrt{\frac{1}{n} \sum_{i=1}^n (x_i - \bar{X})^2} $$

   - Similarly, calculate $\sigma_Y$ for $Y$.

4. The correlation os given by:
     $$
     \rho(X, Y) = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
     $$

### Interpretation of the Correlation Coefficient

- **Range**: The correlation coefficient ranges from $-1$ to $1$.
  - A value close to $1$ indicates a strong positive linear relationship.
  - A value close to $-1$ indicates a strong negative linear relationship.
  - A value close to $0$ indicates no linear relationship.

By following these steps, we can determine the strength and direction of the linear relationship between $X$ and $Y$.

```python
import numpy as np

# Step 1: Generate or define 20 x and y values
# You can replace these with your own data
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(20)  # 20 evenly spaced values from 0 to 19
slope = 2  # Slope of the linear relationship
noise = np.random.normal(0, 8, size=len(x))  # Random noise with mean 0 and standard deviation 2
y = slope * x + noise  # Linear relationship with noise


# Step 2: Calculate the mean of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Step 3: Calculate the covariance
cov_xy = np.mean((x - mean_x) * (y - mean_y))  # Covariance formula

# Step 4: Calculate the standard deviations of x and y
std_x = np.std(x, ddof=0)  # Standard deviation of x
std_y = np.std(y, ddof=0)  # Standard deviation of y

# Step 5: Calculate the correlation
correlation = cov_xy / (std_x * std_y)

# Print the results

print(f"Mean of x: {mean_x:.2f}, Mean of y: {mean_y:.2f}")
print(f"Covariance (X, Y): {cov_xy:.2f}")
print(f"Standard deviation of x: {std_x:.2f}, Standard deviation of y: {std_y:.2f}")
print(f"Correlation (X, Y): {correlation:.2f}")
```

## The Slope of the Best-Fit Line

The slope of the best-fit line through a set of data points is determined by the relationship between the covariance of the variables and the variance of the independent variable ($x$). The formula for the slope is:

$$
\text{slope} = \frac{\text{Cov}(X, Y)}{\sigma_X^2}
$$

### Intuition

- If the covariance is large (positive or negative), it means $X$ and $Y$ have a strong linear relationship, resulting in a steeper slope.
- If the variance of $X$ is small, the slope will be more sensitive to changes in $X$, resulting in a steeper line.
- If the variance of $X$ is large, the slope will be less sensitive to changes in $X$, resulting in a flatter line.



This formula is fundamental in linear regression, where the best-fit line minimizes the sum of squared residuals (differences between observed and predicted $Y$ values).

In the following cell we plot Y versus X and use the upper equation to get slope for a line being the best linear fit:

```python
import numpy as np
import matplotlib.pyplot as plt

slope = cov_xy / (std_x * std_x)

# Create a scatter plot
y_fit = slope * x   # Best-fit line equation

# Step 3: Create the scatter plot and overlay the best-fit line
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Data Points')  # Scatter plot of the data
plt.plot(x, y_fit, color='red', label=f'Best-Fit Line: y = {slope:.2f}x')  # Best-fit line

plt.title("Scatter Plot with Correlation Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```


## Linear Regression and Residual Analysis

## Overview

In this chapter, we will explore linear regression and the analysis of residuals. Linear regression is a powerful tool for modeling relationships between variables, and residual analysis helps us evaluate the quality of the fit.

---

## Linear Regression for Linear and Exponential Data

1. **Linear Data**:
   - For data that follows a linear relationship, we can directly apply linear regression to fit a line of the form:
     $$
     y = \text{slope} \cdot x + \text{intercept}
     $$

2. **Exponential Data**:
   - For data that follows an exponential relationship, such as:
     $$
     y = a \cdot e^{b \cdot x}
     $$
     - We can linearize the data by taking the natural logarithm of $y$:
       $$
       \ln(y) = \ln(a) + b \cdot x
       $$
     - This transforms the problem into a linear regression problem where $\ln(y)$ is the dependent variable.

---

## Residual Analysis

Residuals are the differences between the observed values and the predicted values from the regression model:
$$
\text{Residual} = y_{\text{observed}} - y_{\text{predicted}}
$$

### Key Checks for Residuals:

1. **Linearity**:
   - Residuals should show no clear pattern when plotted against the independent variable ($x$) or the predicted values ($\hat{y}$).
   - A random scatter indicates a good fit.

2. **Normality**:
   - Residuals should follow a normal distribution.
   - Use a **Q-Q plot** (quantile-quantile plot) to visually check normality.


3. **Homoscedasticity**:
   - The spread of residuals should be constant across all levels of the independent variable.
   - Check this by plotting residuals against $x$ or $\hat{y}$ and looking for a consistent spread.

---




```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Generate Linear and Exponential Data
# Linear data
slope_original = 3
intercept_original = 5

x_linear = np.linspace(1, 10, 50)
y_linear = slope_original * x_linear + intercept_original + np.random.normal(0, 6, len(x_linear))  # Linear with noise

plt.figure(figsize=(8, 5))
plt.scatter(x_linear, y_linear, color='lightblue', edgecolor = 'k',label='Data Points')  # Scatter plot of the data

plt.title("New Data")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
# Step 2: Linear Regression for Linear Data
# Use NumPy's polyfit to calculate slope and intercept
slope_linear, intercept_linear = np.polyfit(x_linear, y_linear, 1)
y_pred_linear = slope_linear * x_linear + intercept_linear  # Predicted values for linear data
print('slope = %s' %round(slope_linear,3))
print('intercept = %s' %round(intercept_linear,3))
```

The calculated slope and intercept can be compared to the original values used to generate the data. The differences arise due to the addition of noise, which introduces variability into the data. As a result, the linear model provides an approximation and can only partially reconstruct the original values.

```python
plt.figure(figsize=(8, 5))

# Scatter plot of the data
plt.scatter(x_linear, y_linear, color='lightblue', edgecolor='k', label='Data Points')

# Add the fitted line
plt.plot(x_linear, y_pred_linear, color='red', label=f'Fitted Line: y = {slope_linear:.2f}x + {intercept_linear:.2f}')

# Add title, labels, legend, and grid
plt.title("New Data with Fitted Line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

#### Residual Analysis

```python
# Step 3: Residual Analysis
# Residuals for linear data
residuals_linear = y_linear - y_pred_linear
```

### Homoscedasticity

Plot the residuals against the independent varaible $X$ and check if there is a constant spread.

```python
# Plot Residuals
plt.figure(figsize=(8, 5))

# Residuals for linear data
plt.scatter(x_linear, residuals_linear, color='blue', alpha=0.7)
plt.axhline(0, color='red', linestyle='--')
plt.title("Residuals vs X (Linear Data)")
plt.xlabel("X")
plt.ylabel("Residuals")
```

### Normality

Consider the histogram of the resdiduals and comapre it to a normal distrbution.

```python
# Plot histogram of residuals
plt.figure(figsize=(8, 5))

# Histogram of residuals
plt.hist(residuals_linear, bins=10, density=True, color='lightblue', edgecolor='k', alpha=0.7, label='Residuals')

# Add title, labels, legend, and grid
plt.title("Histogram of Residuals")
plt.xlabel("Residuals")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

### Plotting the Normal Distribution with the Histogram

To evaluate the residuals, they should ideally follow a normal distribution centered around the linear line. To visualize this, we calculate the **sample mean** ($\mu$) and **sample standard deviation** ($\sigma$) of the residuals. Using these values, we define the corresponding normal distribution:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

We then overlay this normal distribution on the histogram of the residuals to assess how closely the residuals align with the expected normal distribution.

```python
# Overlay a normal distribution curve
plt.hist(residuals_linear, bins=10, density=True, color='lightblue', edgecolor='k', alpha=0.7, label='Residuals')


mean_residuals = np.mean(residuals_linear)  # Mean of residuals
std_residuals = np.std(residuals_linear)    # Standard deviation of residuals
x = np.linspace(mean_residuals - 4*std_residuals, mean_residuals + 4*std_residuals, 100)
normal_curve = (1 / (std_residuals * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean_residuals) / std_residuals)**2)
plt.plot(x, normal_curve, color='red', label='Normal Distribution')
plt.title("Histogram of Residuals with Normal Distribution")
plt.xlabel("Residuals")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

### The Q-Q plot



```python
# Manual Q-Q Plot
# Step 1: Sort the data
sorted_data = np.sort(residuals_linear)

# Step 2: Compute theoretical quantiles
n = len(sorted_data)
percentiles = (np.arange(1, n + 1) - 0.5) / n
theoretical_quantiles = stats.norm.ppf(percentiles)

# Step 3: Plot the Q-Q plot manually
plt.scatter(theoretical_quantiles, sorted_data, label="Data Points")
plt.plot(theoretical_quantiles, theoretical_quantiles, color="red", label="Reference Line")  # Diagonal line
plt.title("Q-Q Plot (Manual)")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.legend()
plt.show()
```

#### Interpretation

Straight Line: If the points lie close to the diagonal line, your data is approximately normally distributed.
Deviations:

- S-shaped curve: Indicates heavy tails (data has more extreme values than expected in a normal distribution).
- Inverted S-shaped curve: Indicates light tails (data has fewer extreme values than expected in a normal distribution).
- Systematic deviations: Suggest that the data may follow a different distribution.
